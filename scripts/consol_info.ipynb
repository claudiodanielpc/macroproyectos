```python
#Lectura de pef
pef=pd.read_csv('inputs/opa_ppef_2023.csv',encoding='utf-8')
#Columnas en minusculas
pef.columns=pef.columns.str.lower()
#Eliminar coma en columna 1
pef['clave de cartera']=pef['clave de cartera'].str.replace("'",'')
#Renombrar colunma 1
pef=pef.rename(columns={'clave de cartera':'clave'})
#Renombrar columna 
pef=pef.rename(columns={'descripción de ramo':'desc_ramo'})
pef=pef.rename(columns={'descripción de unidad':'desc_unidad'})
pef=pef.rename(columns={'fecha de inicio':'fecha_inicio'})
pef=pef.rename(columns={'fecha fin':'fecha_fin'})
pef=pef.rename(columns={'total':'costo_total'})
pef
```

```python
#Lectura de proyectos
proyectos=pd.read_csv('inputs/proyectos_mex.csv',encoding='utf-8')
proyectos.columns=proyectos.columns.str.lower()
proyectos=proyectos.rename(columns={'descripción':'desc_proyecto'})
proyectos=proyectos.rename(columns={'estado(s)':'estados'})
proyectos
```

```python
#Pegar pef con proyectos
consolidado=pd.merge(proyectos,pef,on='clave',how='left')
consolidado
```

```python
consolidado.info()
```

```python
#Guardar consolidado en csv
consolidado.to_csv('outputs/consolidado.csv',index=False,encoding='latin-1',errors='ignore')
```
